```python
import os
import shutil
```

```python
QUES_PATH = "assets/Mufti_Folder/MUFTI_QUESTIONS"
ANS_PATH = "assets/Mufti_Folder/MUFTI_ANSWERES"
```

```python
ques_list = os.listdir(QUES_PATH)
ans_list = os.listdir(ANS_PATH)
```

```python
for file in ans_list:
    print(file)
    file_name = os.path.join(ANS_PATH,file)
    shutil.copy(file_name,"assets/audio_classifier_data_3/accepted")
```

```python
for file in ans_list:
    if file in ques_list:
        print(file)
        shutil.copy(os.path.join(QUES_PATH,file),"assets/audio_classifier_data_3/accepted")
```
